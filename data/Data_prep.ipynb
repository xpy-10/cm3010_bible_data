```python
import re
def deconstructColumnNames(stringInput):
    columnList = []
    result =  re.split(r'[〔〕｜＠＋]', stringInput)[1:-1]
    return result 

def returnComplexColumnValue(stringInput, index):
    if deconstructColumnNames(stringInput)[index]==False:
        return ""
    return deconstructColumnNames(stringInput)[index]
```

```python
import numpy as np
```

```python
new_df = df.copy()
```

```python
def addComplexColumnValues(columnNameString, dataFrame):
    namesArray = deconstructColumnNames(columnNameString)
    for index, value in enumerate(namesArray):
        dataFrame[value]=dataFrame[columnNameString].dropna().apply(returnComplexColumnValue, index=index)

multiValuedColumns = [df2, df3, df5]
for value in multiValuedColumns:
    addComplexColumnValues(value.name, new_df)
for value in multiValuedColumns:
    del new_df[value.name]
```

```python
new_df2 = new_df.copy()

#  replaces the value of the multivalued StrongNumbers with two more row entries
def addMultiStrongNumberRow(dataFrame):
    dfIndex = dataFrame.name
    stringEntry = new_df.iloc[dfIndex]['StrongNumber']
    if pd.isna(stringEntry):
        return False
    elif '＋' in stringEntry:
        strongNumberArray = re.split(r'[＋]', stringEntry)
        copiedRow = dataFrame.copy()
        new_df2.at[dfIndex, 'StrongNumber'] = strongNumberArray[0]
        copiedRow['StrongNumber'] = strongNumberArray[1]
        return copiedRow
    else:
        return False
```

```python
# SLOW procedure to concatenate the new rows, 
# calling the function addMuladdMultiStrongNumberRow replaces the value of StrongNumber inplace with the first multivalue
for index, row in new_df.iterrows():
    copiedRow = addMultiStrongNumberRow(row)
    if type(copiedRow) is pd.core.series.Series:
        new_df2 = pd.concat([new_df2.iloc[:index], copiedRow.to_frame().T, new_df2.iloc[index:]]).reset_index(drop=True)
```
